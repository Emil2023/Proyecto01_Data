```python
#Importamos la libreria pandas
import pandas as pd
```

```python
#Cargamos el Dataset
df = pd.read_csv("Datasets\clean_movies_dataset.csv")
```

```python
#Verificamos la info del dataframe
df.info()
```

```python
#Checamos la cantidad de nulos
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
ax.imshow(df.isna(), aspect='auto', interpolation='nearest', cmap='gray')
ax.set_xlabel('Column Number')
ax.set_ylabel('Sample Number')

plt.show()
```

```python
#Realizo un grafico de barras para valores faltantes
import matplotlib.pyplot as plt

# Filtrar las columnas con valores faltantes
mask = df.isna().any(axis=0)

# Calcular el porcentaje de valores faltantes
missing_percentage = df.isna().mean()[mask].sort_values()

# Crear el grfico
fig, ax = plt.subplots(figsize=(15, 4))
ax.bar(missing_percentage.index, missing_percentage.values)
ax.set_title('Porcentaje de valores faltantes por columna')
ax.set_ylabel('Relacin de valores perdidos por columna')

plt.xticks(rotation=90)  # Rotar las etiquetas del eje x si es necesario

plt.show()
```
